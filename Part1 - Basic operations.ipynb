```python
before_1970 = movies["Year"]<1970
movies[before_1970]# Movies before 1970
```

```python
mid_80s = movies["Year"].between(1983,1986)
movies[mid_80s] # Movies made between 1983 and 1986
```

```python
spidermen = movies.index.str.lower().str.contains("spider")
movies[spidermen] #Has spiderman in title
```

```python
movies["Gross"].str.replace(             "$", "", regex = False         ).str.replace(",", "", regex = False)
```

```python
(
     movies["Gross"]            
    .str.replace("$", "", regex = False)            
    .str.replace(",", "", regex = False)            
    .astype(float)
) #Get gross series, remove comma and dollar and make float
```

```python
movies["Gross"]= (
     movies["Gross"]            
    .str.replace("$", "", regex = False)            
    .str.replace(",", "", regex = False)            
    .astype(float)
)
movies["Gross"].mean() #Get mean gross
```

```python
studios = movies.groupby("Studio")
studios["Gross"].count().head()# Group by studios (index is still title) and count no. of movies on tbe list by studios
```

```python
studios["Gross"].count().sort_values(ascending=False).head() #Sort by studios with most films
```

```python
studios["Gross"].sum().head() #Overall gross of each studio
```

```python
studios["Gross"].sum().sort_values(ascending=False).head() #Sort by highest gross
```
